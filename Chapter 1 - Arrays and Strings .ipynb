### Exercise 1.1
**Is Unique**: Implement an algorithm to determine if a string has all unique characters. What if you cannot use additional data structures?

Hints: 
- #44: Try a hash table. 
- #117: Could a bit vector be useful?
- #732: If you were designing this for an NxN board, how might your solution change?

```python
# Initialize Variables
str1 = 'Hello World'
unique_str = 'abcdefg'

# Python Method: ------------------------------------------------------------------------------
def unique_chars_py(str1):
    return len(set(str1)) == len(str1)

# Algorithm Method: ------------------------------------------------------------------------------
def unique_chars_alg(str1):
    vals = [False]*128
    for char in str1:
        if vals[ord(char)]:
            return False
        else:
            vals[ord(char)] = True
    return True

# Algorithm Unique Bit Vector: ------------------------------------------------------------------------------
def unique_char_bitVec(string):
    """Uses bitwise operation instead of extra data structures."""
    # Assuming character set is ASCII (128 characters)
    if len(string) > 128:
        return False
    checker = 0
    for c in string:
        val = ord(c)
        if (checker & (1 << val)) > 0:
            return False
        checker |= 1 << val
    return True

# Algorithm Dictioanry: ------------------------------------------------------------------------------
def unique_char_dict(str1):
    dict_char_counts = {}
    for char in str1:
        if char in dict_char_counts:
            return False
        dict_char_counts[char] = 1
    return True

def unique_char_set(str1):
    characters_seen = set()
    for char in str1:
        if char in characters_seen:
            return False
        characters_seen.add(char)
    return True

# Sorting Method: O(NlogN) ------------------------------------------------------------------------------
def is_unique_chars_sorting(string: str1) -> bool:
    sorted_string = sorted(string)
    last_character = None
    for char in sorted_string:
        if char == last_character:
            return False
        last_character = char
    return True
```

```python
# Test
print('Python way: ')
print('\t ', unique_chars_py(str1))
print('\t ', unique_chars_py(unique_str))

print('\nHash?? Table??: ')
print('\t ', unique_chars_alg(str1))
print('\t ', unique_chars_alg(unique_str))

print('\nBit Vector Way')
print('\t ', unique_char_bitVec(str1))
print('\t ', unique_char_bitVec(unique_str))

print('\nDictionary Way')
print('\t ', unique_char_dict(str1))
print('\t ', unique_char_dict(unique_str))
```

```python
# My Method - Sorting: ------------------------------------------------------------------------------
def str_is_per(str1, str2):
    if len(str1) != len(str2): #print('Diff length\n')
        return False
    
    for i in range(len(str1)):
        if sorted(str1)[i] != sorted(str2)[i]:
            #print(sorted(str[i])) #print(sorted(str2[i]))
            return False
        
    return True

# Python Counter Method: ------------------------------------------------------------------------------
def str_is_per_count(str1, str2):
    if len(str1) != len(str2): #print('Diff length\n')
        return False

    dict = {}
    for char in str1:
        if dict.get(char) == None:
            dict[char] = 1
        else:
            dict[char] += 1

    for char2 in str2:
        if dict.get(char2) == None:
            return False
        else:
            if dict[char2] < 0:
                return False
            dict[char2] = dict[char2]-1
            
    return True


# Python Counter Method: ------------------------------------------------------------------------------
from collections import Counter

def str_is_per_py_count(str1, str2):
    if len(str1) != len(str2): #print('Diff length\n')
        return False

    if Counter(str1) != Counter(str2):
        return False

    #print(Counter(str))
    return True
```

```python
# My Method -- Eliminate by Process of Elimination (Conditions):
def pal_perm(str1):
    dict = {}

    # Use Lowercase
    str1 = str1.lower()

    # Remove or take into a count whitespaces
    space_indexes = [index for index, char in enumerate(str1) if char == " "]
    str1 = str1.replace(' ', '')
    
    # Crear Diccionario:
    for char in str1:
        if dict.get(char) == None:
            dict[char] = 1
        else:
            dict[char] +=1
    
    # Conditions
    # Unique letter Palindrom --
    if len(set(str1)) == 1:
            return True

    # Par -- 
    if len(str1)%2 == 0:
        # -- No Condition: There is a letter repited an odd number of times
        for key in dict.keys():
            if dict[key] % 2 != 0:
                return False

    # Odd --
    else:
        # -- 3 or more letters Palindrom, It is obligatory there is only one letter repeated an odd number of times and all the rest a pair number
        odd_flag = 0
        for key in dict.keys():
            if dict[key] % 2 != 0:
                odd_flag += 1
                if odd_flag > 1:
                    return False

    # Take Spaces into a count
    #space_pal = [' '+char if i in space_indexes else char for i, char in enumerate(str) ]
    #print(space_pal)
    #print(''.join(space_pal))
    #print('Input:', str)
    return True

# Optimized Version: ----------------------------------------------------------------
def can_permute_to_palindrome(input_string):
    # Convert to lowercase and remove spaces (adjust as needed)
    input_string = input_string.lower().replace(" ", "")

    # Create a dictionary to count character frequencies
    char_count = {}
    for char in input_string:
        char_count[char] = char_count.get(char, 0) + 1

    # Count characters with odd frequencies
    odd_count = 0
    for count in char_count.values():
        if count % 2 != 0:
            odd_count += 1
            if odd_count > 1:
                return False

    return True


# ANSWEARS: ----------------------------------------------------------------
def clean_phrase(phrase):
    return [c for c in phrase.lower() if c in string.ascii_lowercase]

# Counter: ----------------------------------------------------------------
def is_palindrome_permutation_pythonic(phrase):
    """function checks if a string is a permutation of a palindrome or not"""
    counter = Counter(clean_phrase(phrase))
    return sum([val % 2 for val in counter.values()]) <= 1
```

```python
def edits_far(input, edit):
    len_diff = len(input) - len(edit)
    # Caso 0 Edits -- lenDiff == 0
    # Caso Replace -- LenDiff == 0
    # Caso Add ------ LenDiff == -1
    # Caso Remove --- LenDiff == 1
    
    if len_diff != 0 and len_diff != 1 and len_diff != -1:
        print('For Length we know its more than one edit --')
        return False

    flag_diff_chars = 0
    flag_same_chars = 0
    # 0 Edits, Replace Case:
    if len_diff == 0:
        for i,c in enumerate(input):
            if(c != edit[i]):
                flag_diff_chars += 1
                if flag_diff_chars > 1:
                    print('Not a Replace, Non Zero Edits --')
                    return False
            flag_same_chars += 1
            
    # Remove Case:
    elif len_diff == 1:
        for i,c in enumerate(input):
            if flag_diff_chars == 0:
                j = i
            else:
                j = i-1
            if input[i] != edit[j]:
                flag_diff_chars += 1
                i += 1
                if flag_diff_chars > 1:
                    print('Not a Substraction Case')
                    return False
            flag_same_chars += 1
            if flag_same_chars == len(input)-1:
                print('Victory Exception -- Last Character Removed')
                return True

    # Add Case:
    elif len_diff == -1:
        for i,c in enumerate(input):
            if c != edit[i]:
                flag_diff_chars += 1
                if c != edit[i+1]:
                    flag_diff_chars += 1
                    print('Not a Substraction Case')
                    return False

    return True

## Recomended Cases
# pale, ple -> true
# pales, pale -> true
# pale, bale -> true
# pale, bake -> false


def one_away_opt(str1, str2):
    len_diff = abs(len(str1) - len(str2))
    
    if len_diff > 1:
        return False

    i, j, edits = 0, 0, 0

    while i < len(str1) and j < len(str2):
        if str1[i] != str2[j]:
            edits += 1
            if len_diff == 1: i, j = (i+1, j) if len(str1) > len(str2) else (i, j+1)
        i, j = i+1, j+1

    return edits <= 1

# Test cases
print(one_away_opt("pale", "ple"))    # True
print(one_away_opt("pales", "pale"))  # True
print(one_away_opt("pale", "bale"))    # True
print(one_away_opt("pale", "bake"))    # False
```

```python
def compress(str1):
    f_char, f_compss, f_total_compss = '', 1, 0
    i = 0
    result = ''
    
    while i < len(str1)-1:
        f_char = str1[i]
        if str1[i] == str1[i+1]:
            f_compss += 1
            f_total_compss += 1
        else:
            result += f_char
            result += str(f_compss)
            f_char = ''
            f_compss = 1
        i += 1

    #Final operation
    result += f_char
    result += str(f_compss)
    
    if f_total_compss >= len(str1)/2:
        print('Nº of comps enough')
        return result

    elif len(result) >= len(str1):
        print('Comp Len Higher')
        return str1

    print('Final Ret')
    return result
    
# Test cases
print(compress('aabcccccaaa'))

# Optimized: ----
def compress(str1):
    compressed = []
    count = 1

    for i in range(1, len(str1)):
        if str1[i] == str1[i - 1]:
            count += 1
        else:
            compressed.append(str1[i - 1] + str(count))
            count = 1

    compressed.append(str1[-1] + str(count))

    result = ''.join(compressed)

    if len(result) < len(str1):
        return result
    else:
        return str1

# Test cases
print(compress('aabcccccaaa'))
```

```python
m = [
    [1,2,3],
    [1,2,3],
    [1,2,3]
]


def rotate_mat(m):
    n = len(m)/4
    return 0

def rotate_matrix(matrix):
    """rotates a matrix 90 degrees clockwise"""
    n = len(matrix)
    for layer in range(n // 2):
        first, last = layer, n - layer - 1
        for i in range(first, last):
            # save top
            top = matrix[layer][i]

            # left -> top
            matrix[layer][i] = matrix[-i - 1][layer]

            # bottom -> left
            matrix[-i - 1][layer] = matrix[-layer - 1][-i - 1]

            # right -> bottom
            matrix[-layer - 1][-i - 1] = matrix[i][-layer - 1]

            # top -> right
            matrix[i][-layer - 1] = top
    return matrix


def rotate_matrix_double_swap(matrix):
    n = len(matrix)
    for i in range(n):
        for j in range(i, n):
            temp = matrix[i][j]
            matrix[i][j] = matrix[j][i]
            matrix[j][i] = temp

    for i in range(n):
        for j in range(int(n / 2)):
            temp = matrix[i][j]
            matrix[i][j] = matrix[i][n - 1 - j]
            matrix[i][n - 1 - j] = temp
    return matrix


def rotate_matrix_pythonic(matrix):
    """rotates a matrix 90 degrees clockwise"""
    n = len(matrix)
    result = [[0] * n for i in range(n)]  # empty list of 0s
    for i, j in zip(range(n), range(n - 1, -1, -1)):  # i counts up, j counts down
        for k in range(n):
            result[k][i] = matrix[j][k]
    return result

def rotate_matrix_pythonic_alternate(matrix):
    """rotates a matrix 90 degrees clockwise"""
    return [list(reversed(row)) for row in zip(*matrix)]


def my_rotate_matrix_double_swap(matrix):
    n = len(matrix)
    # paso 1, trasponer. -- Esto rota la matrix -90º
    # -- (i,j) = (j,i)
    # -- Al Hacer que el bucle interno recorra en range(i,n), gracias a esa I, solo invertimos los elementos que están por encima de la diagonal.
    # ---- Sin ella volveríamos a invertir todo, dejandolo como al comienzo
    for i in range(n):
        for j in range(i,n):
            temp = matrix[i][j]
            matrix[i][j] = matrix[j][i]
            matrix[j][i] = temp
    # Paso 2 Revertir. -- Invertir el orden de las filas -- Rotal 180
    # -- (i,j) = (n-1) == Última Posicion -> (n-1-j) == inverso de j
    # -- Es importante que el bucle interno solo (n/2) revierta la primera mitad de los elementos porque esto ya contempla la segunda mitad
    # ------ Hacerlo completo sería invertir la inversión == Se queda igual
    for i in range(n):
        for j in range(int(n/2)):
            temp = matrix[i][j]
            matrix[i][j] = matrix[i][n-1-j]
            matrix[i][n-1-j] = temp
    
    return matrix

for r in m:
    print(r)
print('---------')

m = [
    [1,2,3],
    [1,2,3],
    [1,2,3]
]
rot = rotate_matrix_double_swap(m)
for r in rot:
    print(r)
print('---------')

m = [
    [1,2,3],
    [1,2,3],
    [1,2,3]
]
rot = my_rotate_matrix_double_swap(m)
for r in rot:
    print(r)
```

```python
import numpy as np

def zero_matrix(matrix):
    n = len(matrix)
    m = len(matrix[0])
    fila = set()
    columna = set()

    for x in range(n):
        for y in range(m):
            if matrix[x][y] == 0:
                fila.add(x)
                columna.add(y)

    print('fila', fila)
    print('col', columna)
    for x in range(n):
        for y in range(m):
            if x in fila or y in columna:
                matrix[x][y] = 0

    return matrix
    

# Funcion que crea vectores vacíos indicando donde esta el $:
def vec(i,n,f):
    if f == 0:
        vec = [0]*n
        vec[i] = '$'
    else:
        vec = [0]*n
    
    return vec
    
### Suponemos que la matriz dada esta compuesta únicamente de ints // o de números positivos tambien sirve
def zero_matrix_zero_memory(matrix):
    temp = 0
    n = len(matrix)
    m = len(matrix[0])
    
    # recorremos la matriz en busca de 0's
    matrix = [['$' if element == 0 else element for element in row] for row in matrix]

    # una vez sabemos donde estan los zeros vamos a eliminar las columnas y filas al rededor suyas:
    for i in range(n):
        for j in range(m):
            if matrix[i][j] == '$':
                matrix[i] = vec(j,m,0)

    tmatrix = np.transpose(matrix)
    for i in range(m):
        for j in range(n):
            if tmatrix[i][j] == '$':                
                tmatrix[i] = vec(i,n,1)

    mat = np.transpose(tmatrix)       
    
    return mat

# Real Solution --------------------------------------------------
def zero_matrix_pythonic(matrix):
    # Replace 0's by 'X'
    matrix = [["X" if x == 0 else x for x in row] for row in matrix]
    indices = []
    # For each (enumerate, row)
    for idx, row in enumerate(matrix):
        # If there is an X in the ROW
        if "X" in row:
            # Recorres la Row y guardas el valor donde hay una X en indices
            indices = indices + [i for i, j in enumerate(row) if j == "X"]
            # Cambias cada row que contenga un 0 con todo ceros
            matrix[idx] = [0] * len(matrix[0])
    # Para cada ROW, para cada Element, 0 si row.index(i )
    matrix = [[0 if row.index(i) in indices else i for i in row] for row in matrix]
    return matrix
```
