### Stack -- LIFO = Last In First Out

```python
Image(url='https://www.simplilearn.com/ice9/free_resources_article_thumb/Stack%20Article%20-%20Soni/working-of-stack.png')
```

```python
class Stack:
    class EmptyStackException(Exception):
        pass
    
    class StackNode:
        def __init__(self, data):
            self.data = data
            self.prev = None
        
    def __init__(self, data=None):
        self.top = None
        if data is not None:
            self.push(data)
        
    def pop(self):
        if self.top is None: raise EmptyStackException('Stack is Empty') 

        item = self.top.data
        self.top = self.top.prev
        return item

    def push(self, data):
        if data == None: raise EmptyStackException('No Data to Add') 

        n = self.StackNode(data)
        n.prev = self.top
        self.top = n
        return

    def peek(self):
        if self.top == None: raise EmptyStackException('Stack is Empty')  
        
        return self.top.data

    def isEmpty(self):
        return self.top == None
```

```python
stack = Stack()
stack.push(3)
stack.peek()
print(stack.top.data)
print('--------------')

stack.push(5)
stack.peek()
print(stack.top.data)
print('--------------')

stack.push(4)
stack.peek()
print(stack.top.data)
print('--------------')

print('pop', stack.pop())
print(stack.top.data)
print('--------------')

print('pop', stack.pop())
print(stack.top.data)
print('--------------')

print('pop', stack.pop())
```

### Queue - FIFO == First In First Out

```python
Image(url='https://cdn.educba.com/academy/wp-content/uploads/2019/12/Queue-in-Data-Structure.jpg')
```

```python
class Queue:
    class EmptyQueueException(Exception):
        pass

    class QueueNode:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.first = None # Head / Front
        self.last = None # Tail / Rear

    def pop(self):
        if self.first is None: raise EmptyQueueExceptio('Empty Queue')

        item = self.first.data
        self.first = self.first.next
        if self.first == None:
            self.last = None
        return item

    def push(self, data):
        if data is None: raise EmptyQueueExceptio('Empty Data')

        n = self.QueueNode(data)
        if self.last is not None:
            self.last.next = n
        self.last = n
        if self.first is None:
            self.first = self.last

    def peek(self):
        if self.last is None: raise EmptyQueueException('EmptyQueue')

        return self.last.data

    def isEmpty():
        return self.first is None
```

```python
queue = Queue()
queue.push(3)
queue.peek()
print(queue.first.data)
print(queue.last.data)
print('--------------')

queue.push(5)
queue.peek()
print(queue.first.data)
print(queue.last.data)
print('--------------')

queue.push(4)
queue.peek()
print(queue.first.data)
print(queue.last.data)
print('--------------')

print('pop', queue.pop())
print(queue.first.data)
print(queue.last.data)
print('--------------')

print('pop', queue.pop())
print(queue.first.data)
print(queue.last.data)
print('--------------')

print('pop', queue.pop())
```

```python
class ArrToStacks:
    def __init__(self, array_size=9):
        self.array = [None] * array_size
        self.stack_sizes = [0] * 3
        self.start_indices = [i * (array_size // 3) for i in range(3)]
        self.end_indices = [(i + 1) * (array_size // 3) - 1 for i in range(3)]
        self.current_stack = 0

    def is_full(self, stack_num):
        return self.stack_sizes[stack_num] == (self.end_indices[stack_num] - self.start_indices[stack_num] + 1)

    def is_empty(self, stack_num):
        return self.stack_sizes[stack_num] == 0

    def push(self, stack_num, data):
        if self.is_full(stack_num):
            self.expand_stack(stack_num)
        
        index = (self.start_indices[stack_num] + self.stack_sizes[stack_num]) % len(self.array)
        self.array[index] = data
        self.stack_sizes[stack_num] += 1

    def pop(self, stack_num):
        if self.is_empty(stack_num):
            raise Exception(f"Stack {stack_num} is empty")
        
        index = (self.start_indices[stack_num] + self.stack_sizes[stack_num] - 1) % len(self.array)
        data = self.array[index]
        self.array[index] = None
        self.stack_sizes[stack_num] -= 1
        return data

    def expand_stack(self, stack_num):
        new_size = len(self.array) * 2
        new_array = [None] * new_size

        for i in range(self.stack_sizes[stack_num]):
            src_index = (self.start_indices[stack_num] + i) % len(self.array)
            new_array[i] = self.array[src_index]

        self.start_indices[stack_num] = 0
        self.end_indices[stack_num] = new_size // 3 - 1

        self.array = new_array

    def switch_stack(self, stack_num):
        if stack_num < 0 or stack_num >= 3:
            raise Exception("Invalid stack number")
        
        self.current_stack = stack_num

# Example usage:
stacks = ArrToStacks(9)  # Create an array that can hold three stacks

stacks.push(0, 1)  # Push 1 onto the first stack
stacks.push(0, 2)  # Push 2 onto the first stack
stacks.push(1, 3)  # Push 3 onto the second stack
stacks.push(2, 4)  # Push 4 onto the third stack

print(stacks.array)  # Output: [1, 2, None, None, 3, None, None, None, 4]
```

**Stack Min:** How would you design a stack which, in addition to push and pop, has a function min which returns the minimum element? Push, pop and min should all operate in 0(1) time.

**Hints:**

- #27: Observe that the minimum element doesn't change very often. It only changes when a smaller element is added, or when the smallest element is popped.
- #59: What if we kept track of extra data at each stack node? What sort of data might make it easier to solve the problem?
- #78: Consider having each node know the minimum of its "substack" (all the elements beneath it, including itself).

```python
class MinStack:
    class EmptyStackException(Exception):
        pass

    class MinStackNode:
        def __init__(self, data = None):
            if data is None: raise EmptyStackException('Data == None')
                
            self.data = data
            self.prev = None
            self.mini = self

    def __init__(self, data=None):
        self.top = None
        self.mini = self
        if data is not None:
            self.push(data)

    def push(self,data):
        if data is None: raise EmptyStackException('Data == None')
        
        n = self.MinStackNode(data)
        if self.top == None or n.mini.data < self.top.mini.data:
            self.mini = n
        else:
            n.mini = self.mini
                
        n.prev = self.top
        self.top = n
        

    def pop(self):
        if self.top is None: raise EmptyStackException('Empty Stack')
            
        if self.top.prev is not None:
            if self.mini.data < self.top.prev.mini.data:
                self.mini = self.top.prev.mini
                
        item = self.top.data
        self.top = self.top.prev
        return item

    def peek(self):
        return self.top.data
        
    def isEmpty(self):
        return self.top == None
        
    def whatsMin(self):
        return self.mini.data
```

Then, after we've retrieved the newest item, putting all the elements back. The issue with this is that doing several pops in a row will require O ( N) work each time. Can we optimize for scenarios where we might do several pops in a row? 

```python
class MyQueue:
    def __init__(self):
        self.stack = Stack()
        self.auxStack = Stack()

    def push(self, data):
        if data is None:
            raise Exceptio('None Data')
            
        self.stack.push(data)

    def pop(self):
        if self.stack.top is None:
            raise Exception('Empty Stack')

        while self.stack.top != None:
            self.auxStack.push(self.stack.pop())

        item = self.auxStack.pop()

        while self.auxStack.top is not None:
            self.stack.push(self.auxStack.pop())

        return item
        
    def peek(self):
        return self.stack.top
    
    def isEmpty(self):
        return self.stack.top is None
    
```
